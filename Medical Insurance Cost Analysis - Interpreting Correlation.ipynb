# Correlation Analysis

## Variables:

- **age**: Represents the age of the individual in years.
- **bmi**: Body Mass Index, a measure of body fat based on height and weight.
- **number_of_children**: Indicates the number of children the individual has.
- **smoker**: A binary value indicating whether the individual is a smoker (1) or non-smoker (0).
- **charges**: Represents the medical insurance costs, assumed to be annual premiums in pounds (£).


## Observations

**bmi vs charges**
- **Positive Trend**: The scatterplot suggests a positive correlation between BMI and insurance charges. As BMI increases, so do the insurance charges.
- **Dense Clusters**: A high density of data points is found between BMI 20 and 30, indicating most individuals fall within this range.
- **Outliers**: Some outliers are seen, mainly in the higher range of BMI and charges, suggesting higher-than-expected insurance costs.

**smoker vs charges**
- **Distinct Separation**: The swarm plot shows clear separation between smokers and non-smokers, with charges for smokers generally higher.
- **Spread and Density**: Smokers and non-smokers have a broader spread, indicating greater variability in charges. Smokers have a more compact distribution.
- **Outliers**: The outliers among smokers are typically at the higher end of the distribution, indicating that smokers often have higher insurance charges.

**age vs charges**
- **Distribution and Pattern**: The scatterplot shows a broad spread of insurance charges across different ages. There is no clear linear trend, suggesting age might not correlate strongly with charges.
- **Clusters**: A significant cluster of data points is observed between ages 25 and 40, suggesting these are the more common age groups.
- **Outliers**: Outliers are mainly found at higher charges.

**number_of_children vs charges**
- **Median Consistency**: The medians for groups with 0 to 3 children are relatively consistent, with medians around 25,000. The group with 4 children has a slightly higher median.
- **Interquartile Range (IQR)**: The IQR varies among groups:
  - Groups with 0, 1, and 2 children have a similar IQR, indicating consistent central spread.
  - Group with 3 children has a slightly wider IQR, suggesting greater variability in charges.
  - Group with 4 children has the narrowest IQR, indicating a more concentrated spread of charges.
- **Outliers**: Groups with 0 and 1 child show several outliers above the upper whisker, indicating a wider spread in insurance charges. Groups with 2 and 3 children have fewer outliers, while the group with 4 children has a single outlier.





## Interpretation
- **bmi vs charges:** The Pearson Correlation Coefficient is 0.71, with a Pearson P-Value close to 0. This indicates a **Strong Correlation** between BMI and insurance charges. The low p-value suggests that this result is **statistically significant** and not due to random chance. The Spearman Correlation Coefficient is slightly higher at 0.73, suggesting a **potential non-linear relationship** or the presence of outliers impacting the Pearson correlation.

- **smoker vs charges:** Boxplots and swarmplots clearly show that **smokers tend to have higher annual insurance charges**. Since smoker status is a nominal categorical variable, Pearson and Spearman correlations are not suitable. The **Point Biserial Correlation** will be used in future analyses to better assess the relationship between smoker status and charges.

- **age vs charges:** The Spearman Correlation Coefficient is 0.07, with a Spearman P-Value of 0.03. This suggests a **very weak correlation** between age and insurance charges. Although the p-value is below 0.05, indicating **statistical significance**, the correlation strength is very low, which may not imply practical relevance.

- **number_of_children vs charges:** The Spearman Correlation Coefficient is 0.05, with a Spearman P-Value of 0.12. This suggests a **very weak correlation** between the number of children and insurance charges. The p-value being above 0.05 means that the observed correlation could be due to chance. Further analysis with a larger dataset or different methods might be needed.

### Limitations
- **Outliers:** Notable outliers were observed for **number_of_children**. These outliers could affect the correlation. Removing them might result in a different p-value or change the correlation strength.
- **Nominal Variables:** A correlation analysis for **smoker vs charges** has not been conducted due to the unsuitability of Pearson and Spearman for nominal categorical data. Future work will employ Point Biserial Correlation for this relationship.


## Python Code
The key variables have been analysed using visual plots and either the Pearson or Spearman Correlation Coefficient. The type of visual plots used varies depending on the relationship between independent and dependent variable. The code has been structured as follows:

0. Import Libraries and Description Statistics.
1. BMI vs Charges (Scatterplot showing linear correlation, Pearson Coeffcient and Heatmap).
2. Smoker vs Charges (Boxplot and Swarmplot)
3. Age vs Charges (Scatterplot indicating no correlation, Spearman Coefficient).
4. Number of Children vs Charges (Boxplot and Swarmplot, Spearman Coefficient).
5. Spearman Heatmap (all variables for simplicity)

#### 0. Import Libraries and Description Statistics

```python
# Import libraries
import pandas  as pd #Data manipulation
import numpy as np #Data manipulation
import matplotlib.pyplot as plt #Visualisation
import seaborn as sns #Visualisation
from scipy.stats import pearsonr
from scipy.stats import spearmanr

data = pd.read_csv("https://raw.githubusercontent.com/fourthrevlxd/cam_dsb/main/insurance.csv")
```

```python
print(data.shape)
data.head()
```

```python
# Exploratory data analysis (EDA).
# Descriptive Statistics.
data.describe(include = "all")
```

#### 1. BMI vs Charges (Pearson)

```python
### Visualisation: BMI vs Charges
plt.figure(figsize=(10, 8))

# Create a scatterplot
plt.scatter(data['bmi'],
            data['charges'],
            c='b')

plt.title('Scatterplot of Charges by BMI')
plt.xlabel('BMI')
plt.ylabel('Insurance Charges per year (£)')

# Show the plot.
plt.show()
```

```python
# Calculate Pearson correlation.
pearson_corr, pearson_p_value = pearsonr(data['bmi'],
                                         data['charges'])

# View the output.
print("\nPearson correlation coefficient:", pearson_corr)
print("Pearson p-value:", pearson_p_value)
```

```python
# Create a smaller dataframe with only BMI and charges
bmi_charges_data = data[['bmi', 'charges']]

# Calculate the Pearson correlation for BMI and charges
corr_bmi_charges = bmi_charges_data.corr(method='pearson')

# Generate the heatmap for the correlation between BMI and charges
plt.figure(figsize=(6, 6))
sns.heatmap(corr_bmi_charges,
            annot=True,
            cmap='coolwarm',
            linewidths=0.5,
            annot_kws={"fontsize": 16, "fontweight": "bold"})

plt.title('Pearson Correlation: BMI and Charges')
plt.show()
```

#### 2. Smoker Status vs Charges (Visual plots only - no Pearson or Spearman)

```python
# Visualisation: Smoker Status vs Charges (Box Plot)
plt.figure(figsize=(10, 8))

# Define the necessary columns
charges = data['charges']
smoker = data['smoker']

# Create a boxplot
smoker_labels = ["Non-smoker", "Smoker"]

plt.boxplot([charges[smoker == 0], charges[smoker == 1]], labels = smoker_labels)

plt.title('Boxplot of Charges by Smoker Status')
plt.xlabel('Smoker status')
plt.ylabel('Insurance Charges per year (£)')

# Show the plot.
plt.show()
```

```python
# Visualisation: Smoker Status vs Charges (Swarm Plot Plot)
plt.figure(figsize=(10, 8))
sns.swarmplot(x=smoker, y=charges, order=[0, 1], palette=["blue", "orange"], hue = smoker, legend=False)

plt.title('Swarm Plot of Charges by Smoker Status')
plt.xlabel('Smoker Status')
plt.ylabel('Insurance Charges per Year (£)')

plt.show()
```

#### 3. Age vs Charges (Spearman)

```python
### Visualisation: Age vs Charges
plt.figure(figsize=(10, 8))

# Create a scatterplot
plt.scatter(data['age'],
            data['charges'],
            c='b')

plt.title('Scatterplot of Charges by Age')
plt.xlabel('Age (years)')
plt.ylabel('Insurance Charges per year (£)')

# Show the plot.
plt.show()
```

```python
# Calculate the Spearman correlation for age
spearman_corr, spearman_p_value = spearmanr(data['age'],
                                            data['charges'])

# View the output.
print("\nSpearman correlation coefficient:", spearman_corr)
print("Spearman p-value:", spearman_p_value)
```

#### 4. Number of Children vs Charges (Spearman)

```python
### Visualisation: Number of Children vs Charges (Boxplot)

plt.figure(figsize=(10, 8))

# Define the necessary columns
charges = data['charges']
number_of_children = data['number_of_children']

# Create a boxplot for number_of_children and charges
number_of_children_labels = np.sort(number_of_children.unique()) # unique values to create labels

plt.boxplot([charges[number_of_children == label] for label in number_of_children_labels],
              labels = number_of_children_labels
  )

plt.title('Boxplot of Charges by Number of Children')
plt.xlabel('Number of Children')
plt.ylabel('Insurance Charges per year (£)')

# Show the plot.
plt.show()
```

```python
### Visualisation: Number of Children vs Charges (Swarm Plot)
plt.figure(figsize=(10, 8))
sns.swarmplot(x=number_of_children,
              y=charges,
              order = number_of_children_labels,
              hue = number_of_children,
              legend = False,
              palette=["blue", "orange", "purple", "green", "red"],
              )

plt.title('Swarm Plot of Charges by Number of Children')
plt.xlabel('Number of Children')
plt.ylabel('Insurance Charges per Year (£)')

plt.show()
```

```python
# Calculate the Spearman correlation for number of children
spearman_corr, spearman_p_value = spearmanr(data['number_of_children'],
                                            data['charges'])

# View the output.
print("\nSpearman correlation coefficient:", spearman_corr)
print("Spearman p-value:", spearman_p_value)
```

#### 5. Spearman correlation matrix (All variables - 'smoker' to be ignored)

```python
# Visualise using a correlation matrix heatmap.
plt.figure(figsize=(10, 8))
corr_matrix_spearman = data.corr(method='spearman')

sns.heatmap(corr_matrix_spearman,
            annot=True,
            cmap='coolwarm',
            linewidths=0.5,
            annot_kws={"fontsize": 16, "fontweight": "bold"})

plt.title('Spearman correlation')

# Show the plot.
plt.show()
```
